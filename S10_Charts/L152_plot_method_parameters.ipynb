```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
bar = pd.read_csv('../data/weather_barcelona.csv', index_col='Date')
rom = pd.read_csv('../data/weather_rome.csv', index_col='Date')
ams = pd.read_csv('../data/weather_amsterdam.csv', index_col='Date')
bar.head()
```

```python
temp_max = pd.DataFrame(index = bar.index)

temp_max['Barcelona'] = bar['TempMax']
temp_max['Rome'] = rom['TempMax']
temp_max['Amsterdam'] = ams['TempMax']

temp_max.head()
```

```python
temp_max.plot()
```

```python
temp_max.plot(figsize=(10,5))
```

```python
temp_max.plot(figsize=(15,5))
```

```python
temp_max.plot(figsize=(15,5), subplots=True)
```

```python
temp_max.plot(figsize=(15,5), use_index=False) # shows row number instead of index
```

```python
temp_max.plot(figsize=(15,5), title='Max Temperature by city') 
```

```python
temp_max.plot(figsize=(15,5), legend=False) 
```

```python
temp_max.plot(figsize=(15,5), logy=True)    #logarythmic
```
