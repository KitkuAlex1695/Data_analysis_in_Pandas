```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
fuel = pd.read_csv('..//data/fuel.csv', 
                   low_memory=False, 
                   usecols=['Vehicle ID','Year','Make','Model',
                            'Class','Fuel Type','Combined MPG (FT1)',
                            'Start Stop Technology'], 
                   index_col='Vehicle ID')
```

```python
fuel.head()
```

```python
listOfCars = ['Renault', 'Toyota', 'Ford']
isInList = fuel['Make'].isin(listOfCars)
isInList.head()
```

```python
fuel[isInList].groupby('Make').count()
```

```python
fuel[isInList].head()
```

```python
isNoStartStopDefined = fuel['Start Stop Technology'].isnull()
isNoStartStopDefined.head()
```

```python
fuel[isNoStartStopDefined].head()
```

```python
isStartStopDefined = fuel['Start Stop Technology'].notnull()
isStartStopDefined
```

```python
#fuel[isStartStopDefined].head()
fuel[isStartStopDefined].groupby('Start Stop Technology').count()
```

```python
mpg50_60 = fuel['Combined MPG (FT1)'].between(50,60)
mpg50_60
```

```python
fuel[mpg50_60].head()
```

```python
x1 = len(fuel[isInList])
x2 = len(fuel[isNoStartStopDefined])
x3 = len(fuel[isStartStopDefined])
x4 = len(fuel[mpg50_60])
```

```python
print(f"""There are {x1} cars within the list ['Renault', 'Toyota', 'Ford']
There are {x2} cars where Start/Stop technology is not determined
There are {x3} cars where Start/Stop technology is determined
There are {x4} cares with MPG value between 50 and 60""")
```
