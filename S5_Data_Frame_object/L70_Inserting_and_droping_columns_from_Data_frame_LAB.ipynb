```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
fuel = pd.read_csv('..//data/fuel.csv', 
                   low_memory=False, 
                   usecols=['Vehicle ID','Year','Make','Model',
                            'Class','Fuel Type','Combined MPG (FT1)'], 
                   index_col='Vehicle ID')
```

```python
fuel.head()
```

```python
mile = 1.609 #km
gallon = 3.78 #l
KPL = mile / gallon
```

```python
fuel['Combined KPL'] = KPL * fuel['Combined MPG (FT1)']
```

```python
fuel.head()
```

```python
fuel.insert(loc=5, column='Liters per km', value=100/fuel['Combined KPL'])
```
