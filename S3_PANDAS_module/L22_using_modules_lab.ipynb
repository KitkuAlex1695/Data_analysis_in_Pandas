```python
months = np.arange(1, 13, 1)
months
```

```python
income = 100 + 3 * months
income
```

```python
plt.plot(months, income, 'go')
```

```python
cost = 50 + 10 * months
cost
```

```python
plt.plot(months, income, 'go', months, cost, 'r^')
```

```python
min = -5
max = 5
step = 0.1
arguments = np.arange(min, max + step, step)

# values = 2 ** arguments

values = [
    pow(2, arguments[x])
    for x in range(len(arguments))
    ]

# print(arguments, values, sep='\n')

plt.plot(arguments, values)
plt.xlabel('Arguments')
plt.ylabel('Values')
```
