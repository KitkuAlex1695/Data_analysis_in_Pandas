```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
mcdonalds = pd.read_csv(filepath_or_buffer  = '../data/mcdonalds.csv',
                        usecols             = ['Category', 'Item', 'Calories', 'TotalFat'])
```

```python
mcdonalds.to_csv(path_or_buf    = 'mc.csv',
                 index          = False)
pd.read_csv('mc.csv')
```

```python
mcdonalds.to_csv(path_or_buf    = 'mc.csv',
                 index          = False,
                 quotechar      = '"')
pd.read_csv('mc.csv')
```

```python
mcdonalds.to_csv(path_or_buf    = 'mc.csv',
                 index          = False,
                 quotechar      = '"',
                 sep            = ',')
pd.read_csv('mc.csv')
```

```python
mcdonalds.to_csv(path_or_buf    = 'mc.csv',
                 index          = False,
                 quotechar      = '"',
                 sep            = ',',
                 columns    = ['Item', 'TotalFat'])
pd.read_csv('mc.csv')
```

```python
mcdonalds.to_json('mc.json')
```
