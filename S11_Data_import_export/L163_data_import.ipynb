```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
url = 'https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/avia_par_pl/?format=TSV&compressed=true'
```

```python
routes = pd.read_csv(url, delimiter='\t', compression='gzip')
routes.head()
```

```python
routes_desc = routes['freq,unit,tra_meas,airp_pr\\TIME_PERIOD'].str.split(',', expand=True)
routes_desc.head()
```

```python
routes_desc.rename({0 : 'freq',
                    1 : 'unit',
                    2 : 'tra_meas',
                    3 : 'airport'},
                    axis = 1,
                    inplace = True)
```

```python
routes_desc = routes_desc.join(routes).drop(axis='columns', labels='freq,unit,tra_meas,airp_pr\\TIME_PERIOD')
```
