```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df_blood = pd.DataFrame({'Group'        :   ['0' ,'0', 'A', 'A', 'B', 'B', 'AB', 'AB'],
                         'Rh'           :   ['+', '-','+', '-','+', '-','+', '-'],
                         'Population'   :   [31, 6, 32, 6, 15, 2, 7, 1]})
```

```python
df_blood
```

```python
df_blood.to_csv(path_or_buf='blood.csv')
```

```python
pd.read_csv('./blood.csv')
```

```python
df_blood.loc[:,"Population"] /= 100
```

```python
df_blood.to_csv(path_or_buf='blood.csv',
                index = False)
```

```python
df_blood.to_csv(path_or_buf = 'blood.csv',
                index = False,
                columns = ['Group', 'Rh'])
```

```python
df_blood.to_csv(path_or_buf = 'blood.csv',
                index = False,
                columns = ['Group', 'Rh'],
                sep = ';')
pd.read_csv('./blood.csv', sep=';')
```

```python
df_blood.to_csv(path_or_buf = 'blood.csv',
                index = False,
                columns = ['Group', 'Rh'],
                sep = ';',
                mode = 'a')
pd.read_csv('./blood.csv', sep=';')
```

```python
df_blood.to_csv(path_or_buf = 'blood.csv',
                index = False,
                columns = ['Group', 'Rh'],
                sep = ';',
                mode = 'a',
                header = False)
pd.read_csv('./blood.csv', sep=';')
```
