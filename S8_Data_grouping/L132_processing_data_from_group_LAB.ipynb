```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('../data/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F',
'Country','State','City'])
df = df[(df['40K'] != '-') & (df['Half'] != '-')]
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head()
```

```python
group_city = df.groupby(by='City')
```

```python
group_city.first()
```

```python
for city in group_city:
    print(city[0])
```

```python
for city in group_city:
    print(city[0], len(city[1]))
```

```python
for city, cityData in group_city:
    print(city, len(cityData))
```

```python
for city, cityData in group_city:
    print(city,
          cityData.nsmallest(1,'TotalSeconds'))
```

```python
the_best_per_city = pd.DataFrame()
```

```python
for city, cityData in group_city:
    the_best_per_city = pd.concat([the_best_per_city,
               cityData.nsmallest(1, 'TotalSeconds')])
```
