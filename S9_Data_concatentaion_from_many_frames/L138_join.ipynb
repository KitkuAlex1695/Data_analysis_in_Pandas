Join method works only on indexes
more complicated examples will be shown later in merge method

args:
- other -> other object
- how ->  left right join
- on -> tells which column in the right frame is to be used as index, if the indexes are different


```python
categories = pd.read_csv('../data/northwind/categories1.csv',
                       usecols=['CategoryID', 'CategoryName', 'Description'])

categories.head()
```

```python
products = pd.read_csv('../data/northwind/products.csv',
                       usecols=['ProductID', 'ProductName', 'UnitPrice', 'CategoryID'])

products.head()
```

```python
categories.set_index('CategoryID', inplace=True)
```

```python
categories
```

```python
products.join(other=categories) # Has no sens -> works only on index
```

```python
products.join(other=categories, on='CategoryID')
```
