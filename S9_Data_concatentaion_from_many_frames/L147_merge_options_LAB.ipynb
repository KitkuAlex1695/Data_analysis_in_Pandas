```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
suppliers = pd.read_csv('../data/northwind/suppliers.csv',
                        usecols=['SupplierID', 'CompanyName', 'City'])
products = pd.read_csv('../data/northwind/products.csv',
                        usecols=['SupplierID', 'ProductName', 'ProductID'])
```

```python
suppliers.merge(right=products, on='SupplierID')
```

```python
suppliers.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(right=products, right_on='SupplierID', left_index=True)
```

```python
products.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(right=products, left_index=True, right_index=True)
```

```python
suppliers.reset_index(inplace=True)
```

```python
suppliers.merge(right=products, left_on='SupplierID', right_index=True)
```

```python
products.reset_index(inplace=True)
```

```python
products.rename({'SupplierID' : 'supplier_id'}, axis='columns', inplace=True)
```

```python
suppliers.merge(products, left_on='SupplierID', right_on='supplier_id')
```
