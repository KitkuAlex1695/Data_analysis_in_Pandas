```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
products = pd.read_csv('../data/northwind/products.csv',
                       usecols=['ProductID', 'ProductName'])
orderedProducts = pd.read_csv('../data/northwind/order-details_del.csv',
                       usecols=['ProductID', 'OrderID', 'UnitPrice', 'Quantity'])
```

```python
products.head()
```

```python
orderedProducts.head()
```

```python
merged = products.merge(right=orderedProducts, 
               on='ProductID', 
               how='left', 
               indicator=True)
merged
```

```python
merged['_merge'].value_counts()
```

```python
merged = products.merge(right=orderedProducts, 
               on='ProductID', 
               how='right', 
               indicator=True)
merged
```

```python
merged['_merge'].value_counts()
```
